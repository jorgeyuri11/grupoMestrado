# Gráfico 


```python
import matplotlib.pyplot as plt
from random import randrange, seed
seed(10)

notas_mat = []
for nota in range (8):
  notas_mat.append(randrange(1,11))


x = list(range(1,9))
y = notas_mat
plt.plot(x,y, marker = 'o')
plt.xlabel ('Provas')
plt.ylabel ('Notas')
plt.title ('Notas de matemática')
plt.show()
```
